# 01 — Exploratory Data Analysis

This notebook provides an initial exploration of the three maternal health datasets used in the
Explainable AI-Based Maternal Triage Decision-Support System.

**Datasets**
| Dataset   | Records | Key additional features |
|-----------|---------|-------------------------|
| FUTH Akure | 149    | Weight, Height |
| Mendeley  | 1206    | BMI, clinical history, Mental Health Status |
| Kaggle/UCI | 1015   | Blood Sugar |


```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.data_loader import load_futh_dataset, load_mendeley_dataset, load_kaggle_dataset, combine_datasets

%matplotlib inline
sns.set_theme(style='whitegrid')
```

## 1. Load Datasets

Place your CSV files in `data/raw/` and update the paths below.

```python
FUTH_PATH      = '../data/raw/futh_dataset.csv'
MENDELEY_PATH  = '../data/raw/mendeley_dataset.csv'
KAGGLE_PATH    = '../data/raw/kaggle_dataset.csv'

df_futh     = load_futh_dataset(FUTH_PATH)
df_mendeley = load_mendeley_dataset(MENDELEY_PATH)
df_kaggle   = load_kaggle_dataset(KAGGLE_PATH)

df = combine_datasets(df_futh, df_mendeley, df_kaggle)
print(f'Combined shape: {df.shape}')
df.head()
```

## 2. Basic Statistics

```python
df.describe()
```

```python
df.info()
```

```python
print('RiskLevel value counts:')
print(df['RiskLevel'].value_counts())
```

## 3. Class Distribution

```python
fig, ax = plt.subplots(figsize=(6, 4))
df['RiskLevel'].value_counts().plot(kind='bar', ax=ax, color=['green', 'orange', 'red'])
ax.set_title('Class Distribution — RiskLevel')
ax.set_xlabel('Risk Level')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

## 4. Missing Value Heatmap

```python
fig, ax = plt.subplots(figsize=(14, 4))
sns.heatmap(df.isnull().T, cbar=False, yticklabels=True, ax=ax)
ax.set_title('Missing Values (white = missing)')
plt.tight_layout()
plt.show()

print('Missing value counts:')
print(df.isnull().sum().sort_values(ascending=False))
```

## 5. Correlation Matrix

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()

fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', ax=ax)
ax.set_title('Pearson Correlation Matrix')
plt.tight_layout()
plt.show()
```
